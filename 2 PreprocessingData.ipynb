# Stopwords

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

# 1. STOPWORDS DASAR
factory = StopWordRemoverFactory()
stopwords = set(factory.get_stop_words())

# 2. TAMBAHAN STOPWORDS
custom_stopwords = {
    'yang','dan','di','ke','dari','ini','itu','jadi','nya',
    'pak','bapak','mbak','lah','sih'
}

stopwords = stopwords.union(custom_stopwords)

# 3. PERTAHANKAN NEGASI
negation_words = {"tidak", "bukan", "jangan"}
stopwords = stopwords - negation_words

# 4. NORMALISASI KATA
replace_dict = {
    'pak': 'prabowo',
    'bapak': 'prabowo'
}

def normalize_tokens(tokens):
    return [replace_dict.get(w, w) for w in tokens]

# 5. STOPWORD REMOVAL
df['tokens_stopword'] = df['Comment'].apply(
    lambda tokens: [
        w.strip().lower() for w in tokens
        if w.strip().lower() not in stopwords
    ]
)

# 6. NORMALISASI SETELAH STOPWORD
df['tokens_stopword'] = df['tokens_stopword'].apply(normalize_tokens)

# 7. GABUNG KEMBALI KE TEKS
df['clean_text'] = df['tokens_stopword'].apply(lambda x: " ".join(x))
```

# Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from tqdm import tqdm
import pandas as pd
import joblib
import os
import re

# 1. CONFIG cache simpan database stemming
CACHE_FILE = "db_stemming.pkl"

# 2. LOAD CACHE
if os.path.exists(CACHE_FILE):
    word_map = joblib.load(CACHE_FILE)
    print(f"✅ Cache loaded ({len(word_map)} kata)")
else:
    word_map = {}
    print("⚠️ Cache baru dibuat")

# 3. INIT STEMMER (sekali)
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# 4. CLEANING TEXT
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z0-9 ]', ' ', text)  # hapus simbol
    text = re.sub(r'\s+', ' ', text).strip()    # hapus spasi ganda
    return text

# 5. STEMMING FUNCTION
def stemming_fast(text):
    text = clean_text(text)
    words = text.split()

    result = []
    for w in words:
        if w in word_map:
            result.append(word_map[w])
        else:
            stem = stemmer.stem(w)
            word_map[w] = stem
            result.append(stem)

    return " ".join(result)


# 6. PROSES DATA
texts = df['tokens_stopword']

hasil = []
for i, text in enumerate(tqdm(texts, desc="Stemming")):
    hasil.append(stemming_fast(text))

    # autosave tiap 500 data
    if i % 500 == 0:
        joblib.dump(word_map, CACHE_FILE)

df['stemmed'] = hasil

# 7. SAVE HASIL
df.to_csv("hasilStemming.csv", index=False)
joblib.dump(word_map, CACHE_FILE)

print("✅ Selesai! Data & cache tersimpan 🚀")
```
